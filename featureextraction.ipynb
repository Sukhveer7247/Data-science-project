```python
def extractepisode(txt):
    episode=""
    signal=False
    for n in txt:
        if(n=="("):
            signal=True
            continue
        if(n==")"):
            break
        if signal:
            episode+=n
    return episode if episode else None
```

```python
episodecount = data["Title"].apply(extractepisode)
```

```python
def extracttime(txt):
    signal=False
    episode=""
    count=0
    for n in txt:
        if(n==")" and count<19):
                signal=True
                continue
        if signal:
                episode+=n
                count+=1
        if(count==19):
                signal=False
                break
    return episode
```

```python
data["Timestamp"]=data["Title"].apply(extracttime)
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

data['Months'] = data['Timestamp'].apply(calculate_total_months)
```
